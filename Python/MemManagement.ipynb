- Nu compara adresele, ci obiectele efectiv

```python
number = 39
print(id(number))

number = number + 1
print(id(number))
```

-   39 nu va mai avea referinta, pentru ca number referenciates 40 deci, obiectul cu valoarea 39 va avea reference count = 0, Garbage Collectorul va dealoca memoria folosita de 39
